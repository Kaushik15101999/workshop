```
def largest(arr,n): 
  
    # Initialize maximum element 
    max = arr[0] 
    max_index = 0
    # Traverse array elements from second 
    # and compare every element with  
    # current max 
    for i in range(1, n): 
        if arr[i] > max: 
            max = arr[i] 
            max_index = i
    return max, max_index 
```
